```python
import pandas as pd
import numpy as np
from datetime import timedelta
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')

# Configuração para o Pandas
pd.set_option('display.max_columns', None)

print("Bibliotecas importadas com sucesso!")
```

```python

# Limpeza e preparação inicial
df_vendas['data_compra'] = pd.to_datetime(df_vendas['data_compra'], dayfirst=True)
df_vendas = df_vendas.dropna(subset=['data_compra', 'id_cliente', 'valor_total_ticket'])
df_vendas = df_vendas[df_vendas['valor_total_ticket'] > 0]
df_vendas.head()
```

```python
# Garantir que os dados estão ordenados por cliente e data da compra
df_ordenado = df_vendas.sort_values(by=['id_cliente', 'data_compra'])

# Calcular a diferença em dias entre cada compra consecutiva do mesmo cliente
df_ordenado['dias_entre_compras'] = df_ordenado.groupby('id_cliente')['data_compra'].diff().dt.days

print("Cálculo de dias entre compras realizado.")
df_ordenado.head()
```

```python
PERIODO_DORMENCIA = 180

# encontrar a data da primeira compra que quebrou uma dormência
primeira_reativacao = df_ordenado[df_ordenado['dias_entre_compras'] >= PERIODO_DORMENCIA].groupby('id_cliente')['data_compra'].min().reset_index()
primeira_reativacao.rename(columns={'data_compra': 'data_reativacao'}, inplace=True)

# encontrar a data da última compra de cada cliente
ultima_compra = df_ordenado.groupby('id_cliente')['data_compra'].max().reset_index()
ultima_compra.rename(columns={'data_compra': 'data_ultima_compra'}, inplace=True)

# juntar as informações apenas para os clientes que adormeceram
df_status = pd.merge(primeira_reativacao, ultima_compra, on='id_cliente')

# classificar os clientes
# se a data da última compra for posterior à data da reativação, ele é "Reativado".
# senão, ele comprou pela última vez quando foi reativado e sumiu de novo, ou nunca foi reativado (aqui simplificamos para "Perdido").
df_status['status'] = np.where(df_status['data_ultima_compra'] > df_status['data_reativacao'], 'Reativado', 'Perdido')


print("Clientes que adormeceram foram classificados com sucesso!")
print(df_status['status'].value_counts())
df_status.head()
```

```python
# preparar a tabela de status com a data limite para cada cliente
df_status_completo = pd.merge(df_status, df_ordenado[['id_cliente', 'data_compra', 'dias_entre_compras']], left_on=['id_cliente', 'data_reativacao'], right_on=['id_cliente', 'data_compra'])
df_status_completo['data_adormeceu'] = df_status_completo['data_reativacao'] - pd.to_timedelta(df_status_completo['dias_entre_compras'], unit='D')
datas_limite = df_status_completo[['id_cliente', 'data_adormeceu']]

# juntar a data limite de cada cliente a TODAS as suas transações
df_com_limite = pd.merge(df_ordenado, datas_limite, on='id_cliente', how='inner')

# filtrar para manter apenas as transações feitas antes da data de adormecer
transacoes_pre_dormencia = df_com_limite[df_com_limite['data_compra'] <= df_com_limite['data_adormeceu']]

# calcular todas as features de uma só vez usando groupby.agg
df_features = transacoes_pre_dormencia.groupby('id_cliente').agg(
    frequencia_pre_dormencia=('id_cliente', 'count'),
    monetario_pre_dormencia=('valor_total_ticket', 'sum'),
    ticket_medio_pre_dormencia=('valor_total_ticket', 'mean')
).reset_index()

# juntar as features com o status (nosso alvo)
df_modelo_final = pd.merge(df_features, df_status[['id_cliente', 'status']], on='id_cliente')

print("\nDataFrame final para modelagem criado")
df_modelo_final.head()
```

```python
# Preparar os dados para o modelo
df_modelo_final['target'] = df_modelo_final['status'].apply(lambda x: 1 if x == 'Reativado' else 0)
df_modelo_final.dropna(inplace=True)

# Selecionar as features (X) e o alvo (y)
X = df_modelo_final[['frequencia_pre_dormencia', 'monetario_pre_dormencia', 'ticket_medio_pre_dormencia']]
y = df_modelo_final['target']

# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Treinar o modelo
model = LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

# Avaliar
y_pred = model.predict(X_test)
print("Relatório de Classificação do Modelo de Reativação")
print(classification_report(y_test, y_pred))

# Interpretação dos Coeficientes!
coeficientes = pd.DataFrame(model.coef_[0], X.columns, columns=['Coeficiente'])
coeficientes['Odds_Ratio'] = np.exp(coeficientes['Coeficiente']).round(2)
coeficientes = coeficientes.sort_values(by='Odds_Ratio', ascending=False)

print("\nPadrão Encontrado: Fatores que Influenciam a Reativação")
print(coeficientes)
```
